## 1.2 Reflections on our Data Wrangling

Our data set is supplied by New York City Dot on https://www1.nyc.gov/html/dot/downloads/pdf/citybench-locations.pdf It is supposed to contain the addresses of all city benches which were installed in NYC in 2019. Unfortunately, the object type of the data set is .pdf which is inappropriate for the pandas package we are using. So we had to take advantage of a .pdf to .xlsx converter: https://smallpdf.com/de/pdf-in-excel. Although we have no gurantee neither for the correctness of the original data set nor for the proper funtionality of the converter, we believe in their reliability. As everywhere in life, even in science, a little bit of trust is necessary to achieve anything. 

Our wrangling shows that several manipulations of the raw data are necessary for answering our research question. The initial data set we are using is not data in the original sense (lat. datum, "the given"). It does not give the answer to our question without further preparation. The data set reveals itself only after a lenthy process of adding, deleting and extracting informationen. So the question arises which moves in data wrangling are legitimate and which moves count as faking the original data.
