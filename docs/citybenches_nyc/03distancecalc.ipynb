## Distance Calculation

```python
#Necessary packages
import osmnx as ox
import networkx as nx
import geopandas 
import itertools
import matplotlib.pyplot as plt
import pandas as pd
import pickle
```

```python
#Choosing a graph for Manhattan
G = ox.graph_from_place('Manhattan, New York City', network_type = 'walk', simplify = False)
```

```python
#Loading the file addressm 
addressm = pd.read_pickle("addressm.pkl")
```

```python
#Nodes of all addresses as a list
Nodes = addressm['Node']
```

```python
#Combine all Nodes 
permutations_object = itertools.permutations(Nodes,2)
```

```python
#Build a permutation object
permutation_list = list(permutations_object)
```

```python
len(permutation_list)
```

## Calculating the distances in packages of 25 0000 entries

```python
distance25k = []
counter = 0
for i,j in permutation_list[:25000]:
    try:
        d = nx.shortest_path_length(G, i, j, weight = 'length')
        counter += 1
        print(counter/25000)
        distance25k.append(d)
    except:
        distance25k.append(" ")
        pass
```

```python
#Saving the result
import pickle
with open('distance25k.txt', 'wb') as fp:
    pickle.dump(distance25k, fp)
```
