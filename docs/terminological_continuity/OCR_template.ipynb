#### 1. pre-processing (i): cut irrelevant material using PyPDF2 package

```python
from PyPDF2 import PdfFileReader, PdfFileWriter
from pathlib import Path

pdf_path = (
    Path.home()
    / "C:/Users/sdbra/Desktop/data/input/raw/Descartes/Meditations/meditations.pdf")
       
input_pdf = PdfFileReader(str(pdf_path))

pdf_writer = PdfFileWriter()
for n in range(15, 19):
    page = input_pdf.getPage(n)
    pdf_writer.addPage(page)
    output_filename = "C:/Users/sdbra/Desktop/data/input/cut/Descartes/Meditations/meditation_one.pdf"

with open(output_filename, 'wb') as out:
                 pdf_writer.write(out)
```

#### 2. pre-processing (ii): turn pdf into jpegs using pdf2image package

```python
from pdf2image import convert_from_path

from pdf2image.exceptions import (
 PDFInfoNotInstalledError,
 PDFPageCountError,
 PDFSyntaxError
)

images = convert_from_path(pdf_path=r'C:\Users\sdbra\Desktop\data\input\cut\Descartes\Meditations\meditation_one.pdf', dpi=300, output_folder=r'C:\Users\sdbra\Desktop\data\output\jpegs\Descartes\Meditations\meditation_one', fmt='jpeg')

# for i, image in enumerate(images):
#    fname = 'sample_Ockham'+str(i)+'.jpeg'
#    image.save(fname, "JPEG")
```

**Version 3.1:** api.AllWordConfidences

```python
%%time

from tesserocr import PyTessBaseAPI

import glob
images = glob.glob(r'C:\Users\sdbra\Desktop\data\output\jpegs\Descartes\Meditations\meditation_one\*.jpg')

# print(images)


with PyTessBaseAPI(path=r'C:\Users\sdbra\Documents\tesserocr-master\tessdata_best-master/.', psm=1, lang='eng') as api:
    for img in images:
        api.SetImageFile(img)
        
        print(api.GetUTF8Text())
        
        print(api.AllWordConfidences())
        
        # print(api.MapWordConfidences)
```
