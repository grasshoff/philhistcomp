### Testing OCR on two extracts of English translations of Ockham's and Suárez' writings

**1. OCR on extract (4 double-pages) of English transl. of Ockham's Quodlibeta Septem** (source: Fredosso & Kelley, Quodlibeta Questions, Yales Uni. Press, pp.256-261, pp.264f.)

```python
%%time

from tesserocr import PyTessBaseAPI

images = [r'C:\Users\sdbra\Documents\tesserocr-master\data\output\sample_Ockham_EN\7208a7fd-d411-420a-9f50-e0e86d70e5b9-1.jpg', r'C:\Users\sdbra\Documents\tesserocr-master\data\output\sample_Ockham_EN\7208a7fd-d411-420a-9f50-e0e86d70e5b9-2.jpg', r'C:\Users\sdbra\Documents\tesserocr-master\data\output\sample_Ockham_EN\7208a7fd-d411-420a-9f50-e0e86d70e5b9-3.jpg', r'C:\Users\sdbra\Documents\tesserocr-master\data\output\sample_Ockham_EN\7208a7fd-d411-420a-9f50-e0e86d70e5b9-4.jpg']

with PyTessBaseAPI(path=r'C:\Users\sdbra\Documents\tesserocr-master\tessdata_best-master/.', psm=1, lang='eng') as api:
    for img in images:
        api.SetImageFile(img)
        
        print(api.GetUTF8Text())
        
        print(api.AllWordConfidences())
```

result: see notebook "OCR_first_steps", test 5.2: works fine, but very slow (>2min for 4 images, however: no comparison yet) and not very handy for processing entire books (i.e. hundreds or even thousands of individual pages/images) since name of every image has to be specified/enumerated in advance
