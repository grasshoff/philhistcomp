### Make a Dataframe out of the Excel-file

```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import json
import numpy as np
```

```python
# make a dataframe from "hcq.xlsx": df_hcq
df_hcq = pd.read_excel("hcq.xlsx",skiprows=[0])
df_hcq.head(3)
```

## Exploit Websites

```python
pubkeys = [
    'title',
    'aff_org_name',
    'researcher_dim_id',
    'researcher_dim_count',
    'journal_title',
    'language',
    'abstract',
    'open_access',
    'publisher',
    'aff_city_name',
    'aff_country_name',
    'doi',
    'pub_date',
    'pub_year',
    'times_cited',
    'altmetric_id',
    'altmetric',
    'authors_full'
    ]
```

```python
# define function getmeta()
def getmeta(x):
    html = ""
    d = ""
    value = ""
    url = x["Dimensions URL"]
    
    if url != None:
        html = requests.get(url).content
        
    soup = BeautifulSoup(html, "lxml")
    datadoc = soup.find("div")
    
    if datadoc == None:
        attr = ""
    else:
        attr = datadoc.get("data-doc")
        if attr:
            d = json.loads(attr)
    
    dicpub = dict()
    
    for i in pubkeys:
        if d != "":
            value = d.get(i, "NaN")
        dicpub[i] = value
        
    return(dicpub)
```

```python
# create a new column "seldict" of dataframe df_hcq
# store metadata in column "seldict" of dataframe df_hcq
df_hcq["seldict"] = df_hcq.apply(getmeta, axis=1)
df_hcq["seldict"].head(3)
```

```python
df_hcq.head(3)
```

```python
lidi = []
def newpub(x):
    di = dict()
    d = x["seldict"]
    di["Publication ID"] = x["Publication ID"]
    di["DOI"] = x["DOI"]
    di["title"] = x["Title"]
    di["authors"] = d["authors_full"]
    di["publisher"] = d["publisher"]
    di["source"] = x["Source title"]
    di["aff_org_name"] = d["aff_org_name"]
    di["aff_country"] = d["aff_country_name"]
    di["pub_date"] = d["pub_date"]
    di["abstract"] = d["abstract"]
    di["openaccess"] = d["open_access"]
    di["di_URL"] = x["Dimensions URL"]
    lidi.append(di)
```

```python
df_hcq.apply(lambda x:newpub(x), axis=1)
df_hcq_Lit = pd.DataFrame(lidi)
df_hcq_Lit.head(3)
```

## Transforming source data

```python
# selcect columns
df_HCQ = df_hcq_Lit[["Publication ID", "title", "abstract"]]
df_HCQ.head()
```

### Remove missing abstracts

We now have a dataframe that contains the title, publication ID and abstract for all selected publictions. As you can see in the cell above, in some rows the abstracts are missing. But the notebook does not register the missing abstracts as missing: Of 19 entries in the column 'abstract' 19 are counted as "non-null" (see the next cell below). Instead it counts empty strings ('') as some kind of abstract - an empty but still existent abstract. It is the next step to get rid of those rows that contain empty strings in the 'abstract' column.  

```python
df_HCQ.info()
```

```python
# Drop rows of 'df_HCQ' that contain in column 'abstract' an empty string ('') instead of an abstract: df_HCQ_dropped

drop_list = []
for i in range(len(df_HCQ)):
    if df_HCQ["abstract"].iloc[i] == '':
        drop_list.append(i)
        
df_HCQ_dropped = df_HCQ.drop(drop_list)
```

```python
# Create new index for 'df_HCQ_dropped'
df_HCQ_dropped.index = [i for i in range(len(df_HCQ_dropped))]
```

```python
df_HCQ_dropped.head()
```

```python
df_HCQ_dropped.info()
```

### Remove weird strings

In the abstracts there can be found strings that are not proper words of the English language, for example "\n" or "&quot;". I suspect that these strings might influence the behaviour of language analysis tools (like 'spaCy'). That is why I'll try to remove them from the abstracts. Strings like "\n" contain escapable characters. To find and remove them, I apply repr() on the abstracts.

```python
# Find weird strings in abstracts an collect strings in a list: weird_strings
weird_strings = []
for i in range(len(df_HCQ_dropped)):
    
    slice_1 = repr(df_HCQ_dropped["abstract"].iloc[i])
    while slice_1.find("\\") >= 0:
        n_1 = slice_1.find("\\")                              
        weird_strings.append(slice_1[n_1:n_1 + 6])
        slice_1 = slice_1[n_1 + 1:]
                                          
    slice_2 = df_HCQ_dropped["abstract"].iloc[i]        
    while slice_2.find("#") >= 0:
        n_2 = slice_2.find("#")
        weird_strings.append(slice_2[n_2:n_2 + 5])
        slice_2 = slice_2[n_2 + 1:]
        
    slice_3 = df_HCQ_dropped["abstract"].iloc[i]        
    while slice_3.find("&quot;") >= 0:
        n_3 = slice_3.find("&quot;")
        weird_strings.append(slice_3[n_3:n_3 + 6])
        slice_3 = slice_3[n_3 + 1:]
```

```python
print(weird_strings)
```

```python
# Remove duplicates from 'weird_strings': weird_strings_without_duplicates
weird_strings_without_duplicates = []

for i in weird_strings:
    if i in weird_strings_without_duplicates:
        continue
    else:
        weird_strings_without_duplicates.append(i)
```

```python
print(weird_strings_without_duplicates)
```

```python
# The list 'weird_strings_without_duplicates' gives an idea of what strings there are in the abstracts that could or should
# be removed. But it still is not a good list. I make some changes by hand and store the resulting list in the variable 
# 'remove_list'.
remove_list = ['\\u2009', '\\n', '\\u2008', '#x27;', '###', '##', '#', '\\xa0', '&quot;']
```

```python
# Remove the strings in 'remove_list' from abstracts. Make a new dataframe that lists the 'clean' abstracst: df_removed
df_removed_list = []

for i in range(len(df_HCQ_dropped)):
    row = []
    row.append(df_HCQ_dropped["Publication ID"].iloc[i])
    row.append(df_HCQ_dropped["title"].iloc[i])
    
    abstract = repr(df_HCQ_dropped["abstract"].iloc[i])
    
    for item in remove_list:
        abstract_removed = abstract.replace(item, " ")
        abstract = abstract_removed
        
    abstract = abstract.replace("& s", "\'s")
    
    row.append(abstract)
    
    df_removed_list.append(row)
    
df_removed = pd.DataFrame(df_removed_list, columns=df_HCQ_dropped.columns)   
```

## Export dataframe to json-file

```python
# export 'df_HCQ_dropped' as json file: df_HCQ.json
df_removed.to_json("df_HCQ.json")
```
