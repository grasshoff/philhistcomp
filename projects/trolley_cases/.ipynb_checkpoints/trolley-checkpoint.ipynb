```python
import pandas as pd
```

```python
#Öffentlich zugängliche Daten aus der Studie von Awad et al. (2019) als .csv Datei importieren
df = pd.read_csv(r'C:\Users\danie\Documents\GitHub\philhistcomp\projects\trolley_cases\assets\Shared_data_responses_demographics.csv')
print(df)
```

```python
#Zur Besseren Nachvollziehbarkeit werden die Antworten der Teilnehmer von '1' und '0' in 'Yes' und 'No' umbenannt
df.loc[df['Outcome'] == 1, 'Outcome'] = 'Yes'
df.loc[df['Outcome'] == 0, 'Outcome'] = 'No'
print(df)
```

```python
#Hier ein Überblick aller deutscher, männlicher bzw. weiblicher Teilnehmer
df.loc[(df['Scenario'] == 'Switch') & (df['survey.gender'] == 'Men') & (df['country_code'] == 'DE')]
```

```python
df.loc[(df['Scenario'] == 'Switch') & (df['survey.gender'] == 'Women') & (df['country_code'] == 'DE')]
```

```python
#Es werden nur Daten angezeigt, die das 'Switch' Szenario betreffen und von männlichen, deutschen Teilnehmern stammen, die im Gedankenexperiment den Hebel
#betätigen würden
df.loc[(df['Scenario'] == 'Switch') & (df['survey.gender'] == 'Men') & (df['country_code'] == 'DE') & (df['Outcome'] == 'Yes')]
```

```python
#Es werden nur Daten angezeigt, die das 'Switch' Szenario betreffen und von weiblichen, deutschen Teilnehmern stammen, die im Gedankenexperiment den Hebel
#betätigen würden
df.loc[(df['Scenario'] == 'Switch') & (df['survey.gender'] == 'Women') & (df['country_code'] == 'DE') & (df['Outcome'] == 'Yes')]
```
