```python
##Numpy, pandas und matplotlib importieren

import numpy as np

import pandas as pd

from pandas import Series, DataFrame

import matplotlib.pyplot as plt
```

```python
#Open Source Daten im csv Format aus der Studie von Awad et al. (2019) importieren
trolley_data = pd.read_csv('Shared_data_responses_demographics.csv')
trolley_data
```

```python
##Durch das Filtern kann anhand der Anzahl der Zeilen festgestellt werden, wieviele der Befragten deutschen Männer im klassischen Trolley-Dilemma ("Switch") den Schalter betätigen würden
data_DE_Men_Yes = trolley_data.loc[(trolley_data['Scenario'] == 'Switch') & (trolley_data['country_code'] == 'DE') & (trolley_data['Outcome'] == 1) & (trolley_data['survey.gender'] == 'Men')]
data_DE_Men_Yes
```

```python
##Mit derselben Funktion kann ermittelt werden, wieviele der Befragten deutschen Männer den Schalter im Gedankenexperiment nicht betätigen würden
data_DE_Men_No = trolley_data.loc[(trolley_data['Scenario'] == 'Switch') & (trolley_data['country_code'] == 'DE') & (trolley_data['Outcome'] == 0) & (trolley_data['survey.gender'] == 'Men')]
data_DE_Men_No
```

```python
##Die Daten werden nochmals nach deutschen Teilnehmerinnen gefiltert
data_DE_Women_Yes = trolley_data.loc[(trolley_data['Scenario'] == 'Switch') & (trolley_data['country_code'] == 'DE') & (trolley_data['Outcome'] == 1) & (trolley_data['survey.gender'] == 'Women')]
data_DE_Women_Yes
```

```python
data_DE_Women_No = trolley_data.loc[(trolley_data['Scenario'] == 'Switch') & (trolley_data['country_code'] == 'DE') & (trolley_data['Outcome'] == 0) & (trolley_data['survey.gender'] == 'Women')]
data_DE_Women_No
```
