# Libraries

```python
import pandas as pd
import re
import spacy
import en_core_web_sm
#nlp=en_core_web_sm.load()
nlp = spacy.load('en_core_web_lg')
```

```python
datawater=pd.read_json("dfwaterpub.json")
alldf=datawater[["PMID","title","abstract"]]
```

```python
raw_text=''
for i in range(2890):
    raw_text+=str(alldf["abstract"].iloc[i])+' '
raw_text1=raw_text[0:1000000]
```

```python
from spacy.lang.en import English

nlp = English()
nlp.add_pipe(nlp.create_pipe('sentencizer'))
doc = nlp(raw_text1)
sentences = [sent.string.strip() for sent in doc.sents]
```

```python
sentences_nlp=[]
for sent in sentences:
    sentences_nlp.append(nlp(sent))
```

```python
import spacy
nlp = spacy.load('en_core_web_lg')

for sent in sentences_nlp[0:10]:
    print(f"*****\nfor {sent}")
    chunks=list(sent.noun_chunks)
    print("noun chunks:",chunks)
```

```python
key_term="water security"
for sent in sentences_nlp:
    chunks=list(sent.noun_chunks)
    for noun in chunks:
        sim=noun.similarity(nlp(key_term))
        print(f"{noun} #### ähnlich: {sim}")
```
