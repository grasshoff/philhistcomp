# Tokenization

```python
# Import spacy
import spacy 
```

```python
# Load library
nlp = spacy.load('en_core_web_sm')
```

```python
# Define variable
bertrand_russell = 'To teach how to live without certainty, and yet without being paralyzed by hesitation, is perhaps the chief thing that philosophy, in our age, can still do for those who study it.'
```

```python
# Tokenize variable
doc = nlp(bertrand_russell)
```

```python
# First seven tokens
doc[0:7]
```

```python
# Return nouns
for chunk in doc[0:7].noun_chunks:
    print(chunk.text)
```

```python
# Defining another variable
marlene_dietrich = 'I still keep a suitcase in Berlin.'
doc1 = nlp(marlene_dietrich)
```

```python
# Recognize entities
for ent in doc1.ents:
    print(ent.text)
    print(ent.label_)
```

```python
# Explain tags and labels of spaCy
spacy.explain('GPE')
```

# Visualizer

```python
# Import displacy
from spacy import displacy
```

```python
# Visualize dependency parse
displacy.render(doc, style='dep', jupyter=True)
```

```python
# Visualize named entities
displacy.render(doc1, style='ent', jupyter=True)
```
